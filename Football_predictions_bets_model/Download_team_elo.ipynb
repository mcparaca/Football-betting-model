```python
import pandas as pd
import numpy as np
```

```python
def get_elo_for_team_on_date(team, match_date, df_elo):
    # Filter by team
    team_elo_data = df_elo[df_elo['Club'] == team]
    # Find the row where the match date is between 'From' and 'To'
    elo_row = team_elo_data[(team_elo_data['From'] <= match_date) & (team_elo_data['To'] >= match_date)]
    if not elo_row.empty:
        return elo_row['Elo'].values[0]
    return None
    
def find_elo_from_team_df(team, match_date, team_dfs):
    # Check if the team has a dataframe
    if team in team_dfs:
        team_df = team_dfs[team]
        # Filter the team_df by the date range
        team_elo = team_df[(team_df['From'] <= match_date) & (team_df['To'] >= match_date)]
        if not team_elo.empty:
            return team_elo['Elo'].values[0]
    return None
```

```python
team_dfs = {}
for i in teams:
    try:
        team_dfs[i] =  pd.read_csv(f'{i}.csv')
    except:
        pass
```

```python
h2h_data = pd.read_csv('h2h_data.csv')
h2h_data['Date'] = pd.to_datetime(h2h_data['Date'])
```

```python

h2h_data['Date'] = pd.to_datetime(h2h_data['Date'])

for team_df in team_dfs.values():
    team_df['From'] = pd.to_datetime(team_df['From'])
    team_df['To'] = pd.to_datetime(team_df['To'])

# Initialize home_elo and away_elo columns
h2h_data['home_elo'] = None
h2h_data['away_elo'] = None

# Function to find the elo of a team from the corresponding team dataframe


# Iterate over each row in df1 to populate home_elo and away_elo columns
for index, row in h2h_data.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    match_date = row['Date']

    # Get home team elo from the corresponding dataframe
    h2h_data.at[index, 'home_elo'] = find_elo_from_team_df(home_team, match_date, team_dfs)
    
    # Get away team elo from the corresponding dataframe
    h2h_data.at[index, 'away_elo'] = find_elo_from_team_df(away_team, match_date, team_dfs)
```

```python

team_elo = pd.read_csv('Brighton.csv')
team_elo['From'] = pd.to_datetime(team_elo['From'])
team_elo['To'] = pd.to_datetime(team_elo['To'])
h2h_data['home_elo'] = h2h_data.apply(lambda row: get_elo_for_team_on_date(row['HomeTeam'], row['Date'], team_elo), axis=1)
h2h_data['away_elo'] = h2h_data.apply(lambda row: get_elo_for_team_on_date(row['AwayTeam'], row['Date'], team_elo), axis=1)
```
