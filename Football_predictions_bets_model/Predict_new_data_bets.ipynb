```python
teams_divs = soup.select('div.n, title')  # Update with the class you showed

home_away_team = []

for div in range(len(teams_divs)):
    txt = teams_divs[div].text.strip()
    if div!=0 and '2024' not in txt and '2025' not in txt:
        c = txt.split(' - ')
        home_away_team.append([c[0],c[1]])
```

```python
eng_tms_dict={'Wolverhampton Wanderers': 'Wolves',
         'Leicester City':'Leicester',
         'Manchester City': 'Man City',
         'Tottenham Hotspur': 'Tottenham',
         'Nottingham Forest': "Nott'm Forest",
         'Manchester United': 'Man United',
         'West Ham United': "West Ham",
         'Newcastle United': 'Newcastle',
         'Brighton & Hove Albion': 'Brighton',
         'AFC Bournemouth': 'Bournemouth',
         'Ipswich Town': 'Ipswich'}
```

```python
home_away_team = np.vectorize(lambda x: eng_tms_dict.get(x, x))(home_away_team)
```

```python
odds_divs = soup.select('div.od div.oo')  # Update with the class you showed
big_odds = []
odds = []
for div in range(len(odds_divs)):
    odd = odds_divs[div].text.strip()
    if len(odd)!=1:
        odds.append(odd)
    if len(odds) == 3:
        big_odds.append(odds)
        odds =[]
```

```python
def convert_odds(season_data):
    bookmaker_margin = []
    for index, row in season_data.iterrows():
        book_mar = (1/row['home_win_odds'] + 1/row['draw_odds'] + 1/row['away_win_odds'])
        home_odds = 1/row['home_win_odds']/book_mar
        draw_odds = 1/row['draw_odds']/book_mar
        away_odds = 1/row['away_win_odds']/book_mar
        season_data.at[index, 'home_win_odds']= home_odds
        season_data.at[index, 'draw_odds']= draw_odds
        season_data.at[index, 'away_win_odds'] = away_odds
    return season_data
```

```python
next_matchday = pd.DataFrame(columns = ['HomeTeam', 'AwayTeam', 'home_win_odds', 'draw_odds', 'away_win_odds'])
next_matchday[['HomeTeam', 'AwayTeam']] = home_away_team
next_matchday[['home_win_odds', 'draw_odds', 'away_win_odds']] =big_odds
next_matchday[['home_win_odds', 'draw_odds', 'away_win_odds']] = next_matchday[['home_win_odds', 'draw_odds', 'away_win_odds']].astype('float')
next_matchday = convert_odds(next_matchday)
```
